<h5>Task 07: String Manipulation and Summarization<h5>
<p><ul><li>Split: Use Python's string manipulation techniques to split the Name column based on a comma, extract the first part, and save it as a new column called SurName.</li>
<li> Group and Count: Use the groupby() function to group by SurName, and apply count() to count
occurrences of each SurName grouped by the FirstName column.</li>
<li>Save: Save the modified dataframe with the new SurName column as titanic2.csv.</li><ul></p>

```python
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv('titanic.csv')

# Split the 'Name' column into 'SurName' and 'FirstName'
# Assuming names are in "Last, First" format
df[['SurName', 'FirstName']] = df['Name'].str.split(',', expand=True)

# Strip extra spaces from 'FirstName' if any
df['FirstName'] = df['FirstName'].str.strip()

# Group by 'SurName' and 'FirstName', then count occurrences
grouped = df.groupby(['SurName', 'FirstName']).size().reset_index(name='Count')

# Display the grouped result
print(grouped)

# Save the modified dataframe with the new SurName column as titanic2.csv
df.to_csv('titanic2.csv', index=False)
```

<h5>Task 10: Boxplot</h5>
<p>Boxplot: Generate a boxplot to compare the distribution of Fare across different Pclass
categories. Preferably, convert Pclass into a categorical variable using the astype('category')
method before creating the plot, and use pandas plot and Matplotlib libraries</p>

```python
# Convert Pclass into a categorical variable 

titanic['Pclass'] = titanic['Pclass'].astype('category')

# Create a boxplot to compare the distribution of Fare across different Pclass categories 

titanic.boxplot(column='Fare', by='Pclass', figsize=(10, 6))

# Add title and axis labels 

plt.title('Fare Distribution by Pclass')
plt.xlabel('Pclass')
```
