```python
# Strategy 1 : Removing rows with missing values
myLogisticRegression(data.dropna())
```

```python
# Strategy 2 : Replacing missing values with mean
data3 = data.copy()
data3 = data3.fillna(data3.mean()) # Fill missing values with mean of the column
myLogisticRegression(data3)
```

```python
# Strategy 2.1 : Replacing missing values with mean using simple imputer
imputer = SimpleImputer(strategy='mean')
data4 = data.copy()
data4[['V7']] = imputer.fit_transform(data4[['V7']])
myLogisticRegression(data4)
```

```python
# Strategy 3 : Replacing missing values with median using simple imputer
data5 = data.copy()
imputer = SimpleImputer(strategy='median')
data5[['V7']] = imputer.fit_transform(data5[['V7']])
myLogisticRegression(data5)
```

```python
# Strategy 4 : Replacing missing values with mode using simple imputer
imputer = SimpleImputer(strategy='most_frequent')
data6 = data.copy()
data6[['V7']] = imputer.fit_transform(data6[['V7']])
myLogisticRegression(data6)
```

```python
# Strategy 5 : Replacing missing values with constant using simple imputer
imputer = SimpleImputer(strategy='constant', fill_value=1)
data7 = data.copy()
data7[['V7']] = imputer.fit_transform(data7[['V7']])
myLogisticRegression(data7)
```

```python
# Strategy 6 : Replacing missing values with KNN
imputer = KNNImputer(n_neighbors=2)
data8 = data.copy()
data8[['V7']] = imputer.fit_transform(data8[['V7']])
myLogisticRegression(data8)
```

```python
# Strategy 7 : Replacing missing values with linear regression wtih other predictors
data9 = data.copy()
train = data9[data9['V7'].notnull()]
test = data9[data9['V7'].isnull()]
print(train.shape) 
print(test.shape)
linReg = LinearRegression()
# model.fit(X,Y)
linReg.fit(train.drop(columns=['V7','V1','V11']),train['V7'])
```

```python
predictions = linReg.predict(test.drop(columns=['V1','V7','V11']))
```

```python
# Rounding predictions
predictions = np.round(predictions,0)
predictions
```

```python
data9.loc[data9['V7'].isnull(),'V7'] = predictions
```

```python
# Strategy 8 : Replacing missing values with linear regression with perturbation
data10 = data.copy()
train = data10[data10['V7'].notnull()]
test = data10[data10['V7'].isnull()]

print(train.shape)
print(test.shape)

linRegPer = LinearRegression()
# model.fit(X,Y)
linRegPer.fit(train.drop(columns = ['V1','V7','V11']), train['V7'])
predictions = linRegPer.predict(test.drop(columns = ['V1','V7','V11']))

# add perturbation to predictions
perturbed_predictions = predictions + np.random.normal(0,0.1,len(predictions))
print(perturbed_predictions)

# Rounding perturbed predictions 
rounded_predictions = np.round(perturbed_predictions,0)
print(rounded_predictions)

# if less than 1 replace with 1 
rounded_predictions[rounded_predictions<1] = 1

# if greater than 10 then replace with 10
rounded_predictions[rounded_predictions>10] = 10

data10.loc[data10['V7'].isnull(), 'V7'] = rounded_predictions
myLogisticRegression(data10)
```
