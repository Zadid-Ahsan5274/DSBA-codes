```python
# Regression with sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Fit linear regression model with all predictors(lm1)
# X = crime_date[["M","Ed","Po1","Po2","LF","M.F","Pop","NW","U1","U2","Wealth","Ineq","Prob","Crime""]]
X = crime_data.drop(columns=['Crime'])
Y = crime_data['Crime']

# Split the data into training(60%), validation(20%), and testing(20%) sets
# train_data, test_data = train_test_split(crime_data, test_size=0.2, random_state=123)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
lm1 = LinearRegression()
```

```python
lm1.fit(X_train, Y_train)
```

```python
# Summary of lm1
print(f"Intercept: {lm1.intercept_}")
print(f"Coefficients: {lm1.coef_}")
```

```python
from sklearn.metrics import root_mean_squared_error
import numpy as np
```

```python
# Predictions on the train set
Y_pred_train = lm1.predict(X_train)

# Predictions on the test set
Y_pred_test = lm1.predict(X_test)

# Calculate the RMSE of the train set
rmse_train = np.sqrt(root_mean_squared_error(Y_train, Y_pred_train))
print("RMSE on train set:", rmse_train)

# Calculate the RMSE of the test set
rmse_test = np.sqrt(root_mean_squared_error(Y_test, Y_pred_test))
print("RMSE on test set:", rmse_test)
```

```python
# Get the R^2 of the model
from sklearn.metrics import r2_score
r2_train = r2_score(Y_train, Y_pred_train)
r2_test = r2_score(Y_test, Y_pred_test)
print("R^2 on training set:", r2_train)
print("R^2 on testing set:", r2_test)
```

```python
# k-fold cross validation
# k = 10
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

lm2 = LinearRegression()
kfold = KFold(n_splits=5, random_state=42, shuffle=True) # shuffle the data before splitting
# we user k-fold for better control over the splitting process, we could simply use cv = k in the cross_val_score function
# print(kfold)

# results = cross_val_score(lm, X, Y, cv=kfold, scoring = 'neg_mean_absolute_error') # sklearn uses negative values
results = cross_val_score(lm2, X, Y, cv=kfold, scoring = 'neg_mean_squared_error') # sklearn uses negative mse

# get rmse
rmse = np.sqrt(-results)
# print rmse
print(rmse)
# get the average rmse
print("Mean RMSE: ", np.sqrt(results.mean()))
```

```python
# now k-fold cross validation for regression tree
# set up k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=1)

# perform k-fold cross validation
scores = cross_val_score(regtree1, X, Y, scoring='neg_mean_squared_error', cv=kf)

# convert the negative MSE scores to positive MSE scores and compute RMSE
rmse_scores = np.sqrt(-scores)
mean_rmse = np.mean(rmse_scores)

print("RMSE scores for each fold: ", rmse_scores)
print("Mean RMSE: ", np.sqrt(mean_rmse))

```

```python
# now random forest regression
from sklearn.ensemble import RandomForestRegressor

# Create a Random Forest Regressor Model
# DecisionTreeRegressor is the base estimator in Random Forest Regressor, so we don't need to specify it explicitly.

# Initialize the Random Forest Regressor Model
randome_forest = RandomForestRegressor(n_estimators=100, random_state=42, max_features=3)
randome_forest.fit(X_train, Y_train)
Y_pred = randome_forest.predict(X_test)
# RMSE
print("RMSE:", np.sqrt(root_mean_squared_error(Y_test, Y_pred)))
```

```python
# now bagging
# Bagging is a machine learning ensemble meta-algorithm that combines multiple base learners (models) to improve generalization performance. The base learners are trained on different subsets of the training data, and their predictions are combined to produce a final prediction. Bagging is often used in situations where a single model is too complex or overfits the training data, and when a small number of models are more effective than a single complex model.
from sklearn.ensemble import BaggingRegressor
bagging_model = BaggingRegressor(n_estimators=100, random_state=0, max_features = X_train.shape[1],n_jobs=-1)
bagging_model.fit(X_train, Y_train)
# RMSE
Y_pred = bagging_model.predict(X_test)
print("RMSE:", np.sqrt(root_mean_squared_error(Y_test, Y_pred)))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
boost_boston = GradientBoostingRegressor(n_estimators=5000, learning_rate=0.001, max_depth=3, random_state=42)
boost_boston.fit(X_train, Y_train)
Y_pred_train = boost_boston.predict(X_train)
Y_pred_test = boost_boston.predict(X_test)
MSE_train = root_mean_squared_error(Y_train, Y_pred_train)
MSE_test = root_mean_squared_error(Y_test, Y_pred_test)
print("MSE on training set: ", MSE_train)
print("MSE on test set: ", MSE_test)
```

```python
X = crime_data[['NW','Po1','Ineq','Prob']]
Y = crime_data['Crime']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
rmse_train = []
rmse_test = []

for k in range(1,21):
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train,Y_train)

    Y_train_pred = knn.predict(X_train)
    Y_test_pred = knn.predict(X_test)

    rmse_train.append(np.sqrt(root_mean_squared_error(Y_train, Y_train_pred)))
    rmse_test.append(np.sqrt(root_mean_squared_error(Y_test, Y_test_pred)))

plt.figure(figsize=(10,5))
plt.plot(range(1,21),rmse_train,label = "Train RMSE")
plt.plot(range(1,21),rmse_test,label = "Test RMSE")
plt.xlabel("k")
plt.ylabel("RMSE")
plt.title("RMSE for train and test data for different k values")
plt.legend()
plt.show()
```

```python
# Support Vector Machine (SVM)
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the SVR model
svr = SVR(kernel='rbf') # kernel can be 'linear', 'poly', 'rbf','sigmoid'

# Fit the SVR model on the training data
svr.fit(X_train_scaled, Y_train)

# Make predictions on the training and testing sets
Y_train_pred = svr.predict(X_train_scaled)
Y_test_pred = svr.predict(X_test_scaled)

# Calculate the mean squared error (MSE) and R^2 on the training and testing sets
mse_train = root_mean_squared_error(Y_train, Y_train_pred)
mse_test = root_mean_squared_error(Y_test, Y_test_pred)
r2_train = r2_score(Y_train, Y_train_pred)
r2_test = r2_score(Y_test, Y_test_pred)

print(f'SVR - Training RMSE: {np.sqrt(mse_train)}, Test RMSE: {np.sqrt(mse_test)}')
print(f'SVR - Training R^2: {r2_train}, Test R^2: {r2_test}')
```
