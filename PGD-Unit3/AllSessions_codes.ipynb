```python
my_anchors = bscontent.find_all('a')
```

```python
# for loop over my_anchors
for anchor in my_anchors:
    # print(anchor.get('href'))
    print(anchor.text)
```

```python
# save anchors, hrefs and texts to a file
list_of_anchors = []
for i in my_anchors:
    my_href = i.get('href')
    my_text = i.text
    list_of_anchors.append((my_href, my_text))
```

```python
list_of_anchors
# convert to a pandas dataframe
import pandas as pd
df = pd.DataFrame(list_of_anchors)
```

```python
df
```

```python
# load into a csv file
df.to_csv('my_anchors.csv',index=False)
```
